## **[2] Histograms & Boxplots for Numerical Features**

---



```python
num_cols = df.select_dtypes(include=np.number).columns
num_cols
```

```python
df[num_cols].hist(bins=15, figsize=(12, 8), color='lightgreen', edgecolor='black')
plt.suptitle("Histograms of all the Numerical Features", fontsize=16)
plt.show()
```

```python
print("\n BOXPLOT of all the Numerical Features :")
for col in num_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[col], color='skyblue')
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[num_cols], orient="h", palette="viridis")
plt.title("Boxplots of All Numerical Features")
plt.tight_layout()
plt.show()
```

## **[3] Pairplot/Correlation Matrix for feature relationships.**

---



## Pairplot

```python
sns.pairplot(df[num_cols], diag_kind='hist')
plt.suptitle('Pairplot of All Numerical Features', y=1.02, fontsize=20)
plt.show()
```

```python
sns.pairplot(df[['Age', 'Fare', 'Pclass', 'Survived']], hue='Survived')
plt.suptitle('Pairplot of Some key Numerical Features', y=1.01, fontsize=15)
plt.show()
```

## Correlation Matrix

```python
corr_matrix = df[num_cols].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

# **[4] Identifying patterns, trends, or anomalies in the data**.

---


```python
# Relationship between Age and Fare
sns.scatterplot(x='Age', y='Fare', data=df)
plt.title("Age vs Fare")
plt.show()

# Outliers in Fare
sns.boxplot(x=df['Fare'], color='lightblue')
plt.title("Fare Outliers")
plt.show()

# Trend in survival rate by gender
sns.barplot(x='Sex', y='Survived', data=df)
plt.title("Survival Rate by Gender")
plt.show()
```

```python
print("\nSurvival rate by Sex:")
print(df.groupby("Sex")["Survived"].mean().round(2))

print("\nSurvival rate by Pclass:")
print(df.groupby("Pclass")["Survived"].mean().round(2))
```
